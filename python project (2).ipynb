```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the retail dataset
df = pd.read_csv("modified_retail_business_data.csv")
```

```python
# Step 2: If Inventory_Days doesn't exist, generate dummy values (for now)
import numpy as np
np.random.seed(42)
df['Inventory_Days'] = np.random.randint(10, 120, size=len(df))
```

```python
# Step 3: Calculate Profit Margin % per sub-category
subcat_df = df.groupby('Sub_Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Inventory_Days': 'mean'
}).reset_index()

subcat_df['Profit_Margin_%'] = round((subcat_df['Profit'] / subcat_df['Sales']) * 100, 2)
```

```python
# Step 4: Correlation Analysis
correlation = subcat_df['Inventory_Days'].corr(subcat_df['Profit_Margin_%'])
print("ðŸ“‰ Correlation between Inventory Days & Profit Margin %:", round(correlation, 2))
```

```python
# Step 5: Visualization
plt.figure(figsize=(10,6))
sns.scatterplot(data=subcat_df, x='Inventory_Days', y='Profit_Margin_%', hue='Sub_Category', s=100)
plt.title("Inventory Days vs Profit Margin % (Sub-Category Level)")
plt.xlabel("Inventory Days")
plt.ylabel("Profit Margin %")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
corr = df[['Inventory_Days', 'Profit', 'Sales']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatterplot
sns.scatterplot(data=df, x='Inventory_Days', y='Profit')
plt.title('Inventory Days vs Profit')
plt.show()
```
